```python
import pandas as pd

data = pd.read_csv('train.csv')
data.head()
```

```python
data.columns
data.isnull().sum()
```

```python
print( data.columns)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
```

```python
data.describe()
```

```python
#Let's choose some features that make sense to be predictive of housing size and build a model around them.
features = ['YearBuilt', 'LotFrontage', 'LotArea','SalePrice']
X = data[features]
X.isnull().sum()
#some of our frontage areas are not present
#lets drop any that aren't there, because that's an easy way of dealing with things.
X = X.dropna()
X.isnull().sum()


y = X.SalePrice
#can't forget to remove sale price from the model. Since that's what we're predicting, we won't have that in the test
del X['SalePrice']
```

```python
#fit the model
linear_model.fit(X, y)
```

```python
#evaluate the model
Eval = linear_model.predict(X)
```

```python
#print out the error to get an idea of how good it was
rmse = ((Eval - y)**2).mean()**.5
print(rmse)
```

```python
#let's add a feature and see if it improves. Number of rooms, above ground seems like another logical predictor.
#check to see if there are any missing numbers
data.TotRmsAbvGrd.isnull().sum()

#redefine X to include rooms
features = ['YearBuilt', 'LotFrontage', 'LotArea','SalePrice','TotRmsAbvGrd']
X2 = data[features]
X2 = X2.dropna()
y2 = X2.SalePrice
del X2['SalePrice']

linear_model2 = LinearRegression()
linear_model2.fit(X2,y2)
Eval2 = linear_model2.predict(X2)
rmse2 = ((Eval2 - y2)**2).mean()**.5
print(rmse2)
```

```python
#our model improved! It still seems pretty bad, but this is a quick model and we don't have time to bother
#with things like "intelligently choosing features" or "engineering our own features".
#let's get something submitted and then build from there 

test = pd.read_csv('test.csv')
```

```python
test.LotFrontage.describe()
```

```python
#this time, we can't just drop the ones we don't like
#instead, we'll fill the missing frontage numbers with the median value.
median_frontage = test.LotFrontage.median()
print(median_frontage)
test.LotFrontage = test.LotFrontage.fillna(median_frontage)
```

```python
#ok, we define our features, make sure that everything is there, and then generate some predictions based on our model.
features = ['YearBuilt', 'LotFrontage', 'LotArea','TotRmsAbvGrd']
X3 = test[features]
X3.isnull().sum()
predictions = linear_model2.predict(X3)

predictions_file = pd.DataFrame({
    'Id': test['Id'],
    'SalePrice': predictions
    })

predictions_file.to_csv('submission.csv', index=False)
```
