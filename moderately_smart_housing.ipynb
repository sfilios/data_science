```python
import pandas as pd

data = pd.read_csv('train.csv')
data.head()
```

```python
#select all the features that are integers. This time, we're going big with features.
features = []
for col in data.columns:
    if data[col].dtype == "int64":
        features.append(col)
        
#set data, but remove SalesPrice
X = data[features]
del X['SalePrice']

#impute features with mean values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer()
Xi = imputer.fit_transform(X)
```

```python
#import the necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor

y = data.SalePrice

#separate for train/test split
train_X, val_X, train_y, val_y = train_test_split(Xi, y, random_state=0)

#define the model and fit the model
forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)

forest_preds = forest_model.predict(val_X)

print(mean_absolute_error(val_y, forest_preds))
```

```python

test = pd.read_csv('test.csv')
```

```python
#exclude the last feature, because it's price data, and doesn't exist in the test set.
features = features[:-1]
X3 = test[features]

#fill in the missing values
X3i = imputer.fit_transform(X3)

#instantiate a new random forest
model = RandomForestRegressor(random_state=1)

#fit the model on our training set, then make predictions
model.fit(Xi, y)
predictions = model.predict(X3i)

#save the predictions as a dataframe and then write it to CSV format
predictions_file = pd.DataFrame({
    'Id': test['Id'],
    'SalePrice': predictions
    })

predictions_file.to_csv('submission_rf.csv', index=False)
```
